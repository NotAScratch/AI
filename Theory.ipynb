## loss Function

```python
import numpy as np

import matplotlib.pyplot as plt

# Define a sample loss function
def loss_function(x):
    return (x - 3) ** 2

# Generate sample data
x = np.linspace(0, 6, 100)
y = loss_function(x)

# Plot the loss function
plt.plot(x, y, label='Loss Function')
plt.xlabel('x')
plt.ylabel('Loss')
plt.title('Loss Function Evolution')
plt.legend()
plt.show()
```
