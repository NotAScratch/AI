```python
df = df.drop(columns=['deck', 'embarked', 'alive', 'class', 'who']) 
```

```python
df.head(30)
```

```python
df.drop_duplicates(inplace=True)
```

```python
sns.boxplot(data=df.select_dtypes(include='number'))
plt.show()
```

```python
num_col = df.select_dtypes(include='number')
q1 = num_col.quantile(0.25)
q3 = num_col.quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

df[num_col.columns] = num_col.clip(lower = lower_bound, upper = upper_bound, axis = 1)
```

```python
df.head(30)
```

```python
cols = ['sex', 'embark_town', 'alone', 'pclass', 'adult_male']
df = pd.get_dummies(df, columns=cols, drop_first=True, dtype=int)
```

```python
import sklearn as sk
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
df.head()
```

```python
scaler = StandardScaler()
scaling_cols = ['age', 'fare']
df[scaling_cols] = scaler.fit_transform(df[scaling_cols])
```

```python
x = df.drop(columns='survived')
y = df['survived']

x-train, x_test, y_train, y_test = sk.model_selection.train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
```
